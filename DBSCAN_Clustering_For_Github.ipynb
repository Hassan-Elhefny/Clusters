# Creating Dataset using [sklearn.datasets.make_blobs](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html) library


```python
#import libraries to create dataset
from sklearn.datasets import make_blobs ,make_circles, make_s_curve
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#parameters of datasets
nSamples = 500  #number of samples or records
nFeatures = 2  #number of features
rs = 1024  #random state
```

```python
#Create three datasets
dataset_1 = make_blobs(nSamples, nFeatures, cluster_std=0.65, random_state=rs)[0] #to get the data without label (nfeatures) = 2
dataset_2 = make_circles(nSamples, noise=0.05, factor=0.5, random_state=rs)[0]    #to get the data without labels = 2
dataset_3 = make_s_curve(nSamples, noise=0.15, random_state=rs)[0][:, [0, 2]]    #to just get the first 2 features to make it easy to plot but this method return 3 features
```

```python
#plot the dataset_1
print(dataset_1.shape)
plt.scatter(dataset_1[:, 0], dataset_1[:, 1])
plt.show()
```

```python
#plot the dataset_2
print(dataset_2.shape)
plt.scatter(dataset_2[:, 0], dataset_2[:, 1])
plt.show()
```

```python
#plot the dataset_3
print(dataset_3.shape)
plt.scatter(dataset_3[:, 0], dataset_3[:, 1])
plt.show()
```

# DBSCAN (Density-Based Spatial Clustering of Applications with Noise)
image_2022-09-01_111110426.png
DBSCAN groups data points together based on the distance metric. It follows the criterion for a minimum number of data points. It can discover clusters of different shapes and sizes from a large amount of data, which is containing noise and outliers.It takes two parameters – eps and minimum points. Eps indicates how close the data points should be to be considered as neighbors. The criterion for minimum points should be completed to consider that region as a dense region.

```python
#import important libraries
from sklearn.cluster import DBSCAN
import numpy as np
```

```python
#Building Blob DBSCAN Model
name = "Blobs DBSCAN"
eps = 0.6
blob_model = DBSCAN(eps=eps, min_samples=1)
predicted_blob_labels = blob_model.fit_predict(dataset_1)

#print shape od Dataset
print(dataset_1.shape)

#Plot blob DBSCAN
plt.figure(figsize=(10,10))
plt.scatter(dataset_1[:, 0], dataset_1[:, 1], c=predicted_blob_labels)
plt.title(name)
plt.xlabel("X")
plt.ylabel("Y")
```

```python
#print predicted blob model and unique values
print(np.unique(predicted_blob_labels, return_counts=True))
print()
print(predicted_blob_labels)
```

```python
#Building Circle DBSCAN Model
name = "Circle DBSCAN"
eps = 0.2
circle_model = DBSCAN(eps=eps, min_samples=1)
predicted_circle_labels = circle_model.fit_predict(dataset_2)

#print shape od Dataset
print(dataset_2.shape)

#Plot blob DBSCAN
plt.figure(figsize=(10,10))
plt.scatter(dataset_2[:, 0], dataset_2[:, 1], c=predicted_circle_labels)
plt.title(name)
plt.xlabel("X")
plt.ylabel("Y")
```

```python
#print predicted circle model and unique values
print(np.unique(predicted_circle_labels, return_counts=True))
print()
print(predicted_circle_labels)
```

```python
#Building S DBSCAN Model
name = "S DBSCAN"
eps = 0.5
s_model = DBSCAN(eps=eps, min_samples=1)
predicted_s_labels = s_model.fit_predict(dataset_3)

#print shape od Dataset
print(dataset_3.shape)

#Plot blob DBSCAN
plt.figure(figsize=(10,10))
plt.scatter(dataset_3[:, 0], dataset_3[:, 1], c=predicted_s_labels)
plt.title(name)
plt.xlabel("X")
plt.ylabel("Y")
```

```python
#print predicted S model and unique values
print(np.unique(predicted_s_labels, return_counts=True))
print()
print(predicted_s_labels)
```
